```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.set_printoptions(suppress = True)
df = pd.read_csv('https://raw.githubusercontent.com/cindysame179/Data-Analytics-Work/main/diamonds.csv')
df.rename(columns = {'Unnamed: 0':'Index'}, inplace = True)
#dropping the diamonds with dimensions valued at 0 as they do not make sense
df.drop(df[(df['x']== 0) | (df['y']== 0) | (df['z'] == 0)].index, inplace = True)
#checks for nulls
df.isnull().values.any()
```

#**Data Attributes**

**Price** - Price of the Diamond in US dollars, with the cheapest in the dataset being 326 dollars and the most expensive being 18,823 dollars. This value is represented as an integer in the table.

**Carat** - Weight of the diamond ranging from 0.2 to 5.01 carats. One carat is equivalent to one fifth of a gram or 200 milligrams. This value is represented as an float in the table.

**Cut** - Quality of how the diamond was cut; given a possible rating of Fair, Good, Very Good, Premium, or Ideal. This value is represented as an string in the table.

**Color** - Color of the diamond. Typically rated from D (colorless) to Z (yellow/brown); however, this data set only contains diamonds rated on a scale from D to J (nearly colorless diamonds). This value is represented as an string in the table. 
<img src="https://3kqo2r1c2kzy3oyiqn2r1gjb-wpengine.netdna-ssl.com/wp-content/uploads/2020/11/4c1-1.jpg" alt="Diamond Color Scale"> <sub>*Picture taken from the American Gem Society*</sub>

**Clarity** - a measurement of how clear a diamond is and how many imperfections there are within the diamond. Data set contains diamonds that have the rating LF(best), VVS1, VVS2, VS1, VS2, SI1, SI2, and I1(worst). This value is represented as an string in the table.

<img src="https://cdn.shopify.com/s/files/1/0370/7551/7485/files/clarity.jpg" alt="Diamond Clarity Scale"> <sub>*Picture taken from Joseph Schubach Jewelers*</sub>

**x** - length of the diamond in millimeters. This value is represented as an float in the table.

**y** - width of the diamond in millimeters. This value is represented as an float in the table.

**z** - height of the diamond in millimeters. This value is represented as an float in the table.

**Depth** - Total depth percentage of a diamond. Gives an idea of how brilliantly a diamond shines as if a diamond is too shallow or too deep, light will begin to disperse through the side or bottom instead of the top. This value was obtained in this dataset through the equation:$$TotalDepthPercentage = \frac{z}{mean(x,y)} = \frac{2*z}{x+y}$$ and contained the values from 43% - 79%. This value is represented as an float in the table.

**Table** - width of the largest facet on top of the diamond. Values range from 43 to 95. This value is represented as an float in the table.

#**What proportion of diamonds are between .30 and 1.08 carats?**

```python
#get the number of diamonds with carats >.3 and <1.08, divide that by the total number of diamonds in the data set and multiply by 100 to get the percentage
proportion = df[(df['carat'] > .3) & (df['carat'] < 1.08)].shape[0]/df.shape[0]*100
print(f'Approximately {"%.2f" % proportion}% of diamonds in this data set are between .30 and 1.08 carats')
```

#**How many of the diamonds have equal x and y dimensions?**

```python
#create a new dataframe with only rows with equal x and y dimensions and count up those rows to get the total amount of diamonds with this property
equalxy = df[df['x'] == df['y']].shape[0]
print(f'There are {equalxy} diamonds that have equal x and y dimensions')
```

#**How many of the diamonds have a carat value that is less than the mean carat value?**


```python
#finds the mean carat value
mean = np.mean([df['carat']])
#create dataframe containing all the diamonds with lower carat values and count up the rows to get the total number of diamonds with this property
ltm = df[df['carat'] < mean].shape[0]
print(f'There are {ltm} diamonds that have a carat value that is less than the mean carat value')
```

#**How many diamonds have a Premium cut or better? Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal**

```python
#create a dataframe containing only diamonds with cut values of premium and ideal. Then count up the rows for the total amount
PremAndIdeal = df[(df['cut'] == 'Premium') | (df['cut'] == 'Ideal')].shape[0]
print(f'There are {PremAndIdeal} diamonds that have a Premium cut or better')
```

#**Which diamond has the highest price per carat? What is its value?**

```python
#get the price per carat of each diamond
ppcdf = df['price']/df['carat']
#find the max value
max = ppcdf.max()
#find the diamond number with the same value
maxppcdf = (df[df['price']/df['carat'] == max])
index = maxppcdf.iloc[0]['Index']
print(f'Diamond {index} has the highest price per carat valued at {max}')
```

#**Using both Matplolib and Seaborn, make a scatter plot of price vs. carat. What can we say about the relationship between those two attributes?**

```python
plt.figure(1)
plt.subplot(121)
plt.scatter(df['carat'], df['price'])
plt.title('Price vs Carat Matplolib ScatterPlot')
plt.subplot(122)
sns.scatterplot(x = 'carat', y = 'price', data = df)
plt.title('Price vs Carat Seaborn ScatterPlot')
plt.subplots_adjust(left= .1, right = 2)
                  
```

From the scatterplots generated by the dataset, we can see a general relationship between price and carat. This relationship being that if the carat value is higher, the price will also be higher. This is because in both scatter plots, the graph going in an upwards diagonal direction from left to right.
